```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from Classes import DataProcessor
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
df = pd.read_csv("Quality_of_Life.csv")
df
```

```python
df = DataProcessor(df).process_all()

df.head()
```

```python
df.info()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

sns.histplot(df, color='blue', ax=axes[0], kde=True, x="Valor do Clima")

sns.histplot(df, color='red', ax=axes[1], kde=True, x='Valor do Custo de Vida')

plt.show()
```

```python
df = DataProcessor(df).remove_columns_nocorr()
```

```python
df.corr()
```

```python
sns.regplot(data=df, x="Valor do Custo de Vida", y="Valor do Clima", 
            scatter_kws={'color': 'blue'}, line_kws={'color': 'red', 'linewidth': 2})
```

```python
X = df['Valor do Clima']
y = df['Valor do Custo de Vida']

X = sm.add_constant(X)

modelo = sm.OLS(y, X)
```

```python
resultado = modelo.fit()
print(resultado.summary())
```
